```python
#Importamos los módulos necesarios:
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Definimos las funciones que se utilizaran para calcular ambos métodos
def  mesh(a, b, Nt):
    """
    Regresa el espaciamiento de malla, con un punto de inicio (a) y un punto de final (b), con Nt número de espaciamiento
    """
    ht = (b-a) / Nt
    return  ht

def  exactSolution(t, y0, lam):
    """
    Regresa la solución exacta al problema con el dominio (t), la condición inicial y0, y el valor de lambda (lam)
    """
    return  y0 /(y0+(1-y0)*np.exp(-lam * t))
```

#### Para λ= 10, y0= 0.01 y Nt= 4,16,64.

```python

Nt4 = 4
Nt16 = 16
Nt64= 64

Tmax   = 1

ht4 = mesh(0, Tmax , Nt4)
ht16 = mesh(0, Tmax , Nt16)
ht64 = mesh(0, Tmax , Nt64)

y0 = 0.01
lam = 10

t4   = np.linspace(0, Tmax , Nt4+1)
t16   = np.linspace(0, Tmax , Nt16+1)
t64   = np.linspace(0, Tmax , Nt64+1)

yf4 = np.zeros(Nt4+1)
yf16 = np.zeros(Nt16+1)
yf64 = np.zeros(Nt64+1)

#Para Nt=4
yf4[0] = y0
for i in range(0,Nt4):
    yf4[i+1]=yf4[i]*(1+ht4*lam*(1-yf4[i]))

#Para Nt=16
yf16[0] = y0
for i in range(0,Nt16):
    yf16[i+1]=yf16[i]*(1+ht16*lam*(1-yf16[i]))

#Para Nt=64
yf64[0] = y0
for i in range(0,Nt64):
    yf64[i+1]=yf64[i]*(1+ht64*lam*(1-yf64[i]))
    
#Solucion exacta
tl = np.linspace(0, Tmax , 100)
y_exacta = exactSolution(tl, y0 , lam)

y_exac_p_4 = exactSolution(t4, y0, lam)
y_exac_p_16 = exactSolution(t16, y0, lam)
y_exac_p_64 = exactSolution(t64, y0, lam)

#Calculo del error
norma_error_f_4 = np.linalg.norm(yf4 - y_exac_p_4 ,2)
norma_error_f_16 = np.linalg.norm(yf16 - y_exac_p_16 ,2)
norma_error_f_64 = np.linalg.norm(yf64 - y_exac_p_64 ,2)

#Graficos
Ecuacion = '$y(t) =\dfrac {y_0} {y_0+(1+y_0)e^{-\lambda t}}$'
fig , (ax1) = plt.subplots(1,1)
fig.suptitle('Función Logística',x=0.2)
ax1.plot(tl, y_exacta , 'k-', lw=3, label='Sol. Exacta')
ax1.plot(t4, yf4 ,'o--', label='Nt={}. Error={:03.5f}'.format(Nt4,norma_error_f_4))
ax1.plot(t16, yf16 , 'gv--', label='Nt={}. Error={:03.5f}'.format(Nt16,norma_error_f_16))
ax1.plot(t64, yf64 , 'm*--', label='Nt={}. Error={:03.5f}'.format(Nt64,norma_error_f_64))
ax1.set_title(Ecuacion, fontsize =12,  color='blue')
ax1.set_xlabel('$t$')
ax1.set_ylabel('$y(t)$')
ax1.legend(loc='upper left', ncol=1,  framealpha =0.75,  fancybox=True , fontsize =10)
ax1.grid()
plt.savefig('logistica_varios.pdf')
plt.show()
```

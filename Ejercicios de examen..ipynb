```python
#Inciso a:
#Funcion:
f=lambda x: 1 - np.exp(x) + (np.exp(1) - 1) * (np.sin(((np.pi * x) / 2)))

#Su derivada es:

df=lambda x: ( ( (np.exp(1) - 1) * np.pi * np.cos((np.pi* x)/2)) / 2) - np.exp(x)

# Dominio de la función
xmin = -1,1
xmax = 2.1
x = np.linspace(xmin, xmax, 100)

#Gráfica de la función
plt.plot(x,df(x), 'r-', lw=2)
plt.title('Inciso $a$')
plt.grid()
plt.show()

#Comprobación de Teorema de valor medio
vm=(f(1)-f(0))/(1-0)
print(vm)
```
